```python
# Step 1: Data storage
expenses = []
# New dictionary to store { "Category": budget_amount }
budgets = {}

# Step 2: Function to set a budget
def set_budget():
    category = input("Enter category to set budget for (e.g., Food, Travel): ").capitalize()
    try:
        amount = float(input(f"Enter budget limit for {category}: ").replace('$', ''))
        budgets[category] = amount
        print(f"Budget for {category} set to ${amount:.2f}\n")
    except ValueError:
        print("Invalid amount. Budget not set.")

# Step 3: Function to add an expense
def add_expense():
    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category (Food/Travel/Other): ").capitalize()
    while True:
        try:
            amount_str = input("Enter amount: ")
            amount = float(amount_str.replace('$', ''))
            break
        except ValueError:
            print("Invalid amount. Please enter a numerical value.")

    description = input("Enter description: ")

    expense = {
        "date": date,
        "category": category,
        "amount": amount,
        "description": description
    }
    expenses.append(expense)
    print("Expense added successfully!\n")

# Step 4: Function to view expenses with Budget Logic
def view_expenses():
    if not expenses:
        print("No expenses recorded yet.\n")
        return

    # Categorically add functionality to cumulatively add expenses
    category_totals = {}
    for exp in expenses:
        cat = exp['category']
        category_totals[cat] = category_totals.get(cat, 0) + exp['amount']

    print("\n--- Your Expenses & Budget Status ---")
    print(f"{'Date':<12} | {'Category':<10} | {'Amount':<8} | {'Budget':<8} | {'Remaining':<8}")
    print("-" * 60)

    for exp in expenses:
        cat = exp['category']
        budget = budgets.get(cat, 0.0)
        total_spent_in_cat = category_totals[cat]
        remaining = budget - total_spent_in_cat

        # Displaying the specific expense line with the category's overall budget status
        print(f"{exp['date']:<12} | {cat:<10} | {exp['amount']:<8.2f} | {budget:<8.2f} | {remaining:<8.2f}")
    print()

# Step 5: Simple menu
while True:
    print("1. Add Expense")
    print("2. Set/Update Category Budget")
    print("3. View Expenses & Remaining Budgets")
    print("4. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        add_expense()
    elif choice == '2':
        set_budget()
    elif choice == '3':
        view_expenses()
    elif choice == '4':
        print("Goodbye!")
        break
    else:
        print("Invalid choice, try again.")
```
